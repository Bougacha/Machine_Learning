```python
#Create a function to plot confusion matrix

def plot_confusion_matrix(y_test,pred_test):
  cm=confusion_matrix(y_test,pred_test)
  plt.clf()
  plt.imshow(cm , interpolation='nearest',cmap=plt.cm.Accent)
  categoryNames=['Non-diabet','diabet']
  plt.title('Confusion Matrix - Test Data')
  plt.ylabel('True label')
  plt.xlabel('predict label')
  ticks=np.arange(len(categoryNames))
  plt.xticks(ticks,categoryNames,rotation=45)
  plt.yticks(ticks,categoryNames)
  s=[['TN','FP'],['FN','TP']]
  for i in range(2):
    for j in range(2):
      plt.text(j,i,str(s[i][j]) + "=" + str(cm[i][j]) , fontsize=12)
  plt.show()
```

```python
def logregression (Results,Methodology,x_train, x_test, y_train, y_test):

  from sklearn import linear_model
  from sklearn.model_selection import KFold
  from sklearn.metrics import roc_auc_score

  num_C = list(np.power(10.0, np.arange(-10, 10)))
  cv_num = KFold(n_splits=10, shuffle=True, random_state=42)

  searchCV_l2 = linear_model.LogisticRegressionCV(
          Cs= num_C
          ,penalty='l2'
          ,scoring='roc_auc'
          ,cv=cv_num
          ,random_state=42
          ,max_iter=10000
          ,fit_intercept=True
          ,solver='newton-cg'
          ,tol=10 )

  searchCV_l1 = linear_model.LogisticRegressionCV(
          Cs=num_C
          ,penalty='l1'
          ,scoring='roc_auc'
          ,cv=cv_num
          ,random_state=42
          ,max_iter=10000
          ,fit_intercept=True
          ,solver='liblinear'
          ,tol=10 )

  searchCV_l1.fit(x_train, y_train)
  searchCV_l2.fit(x_train, y_train)
  print ('Max auc_roc for l1:', searchCV_l1.scores_[1].mean(axis=0).max())
  print ('Max auc_roc for l2:', searchCV_l2.scores_[1].mean(axis=0).max())

  print("***********Parameters for l1 regularisations************")
  print(searchCV_l1.coef_)
  print(searchCV_l1.intercept_)
  print(searchCV_l1.scores_)

  print("***********Parameters for l2 regularisations*************")
  print(searchCV_l2.coef_)
  print(searchCV_l2.intercept_)
  print(searchCV_l2.scores_)

#find predicted values
  y_pred_l1 = searchCV_l1.predict(x_test)
  y_pred_l2 = searchCV_l2.predict(x_test)

#find predicted probabilities
  y_pred_probs_l1 = searchCV_l1.predict_proba(x_test)[:,1]
  y_pred_probs_l2 = searchCV_l2.predict_proba(x_test)[:,1]

# Accuaracy of L2/L1 models
  f1_score_l2 = metrics.f1_score(y_pred=y_pred_l2, y_true=y_test)
  f1_score_l1 = metrics.f1_score(y_pred=y_pred_l1, y_true=y_test)

  print("f1_score of Logistic model with l2 regularisation -------" , f1_score_l2)
  print("*******Confusion Matrix********")
  plot_confusion_matrix(y_test, y_pred_l2)
  print("**********classification Report*********")
  print(classification_report(y_test, y_pred_l2))

  print("f1_score of Logistic model with l1 regularisation -------" , f1_score_l1)
  print("*******Confusion Matrix********")
  plot_confusion_matrix(y_test, y_pred_l1)
  print("**********classification Report*********")
  print(classification_report(y_test, y_pred_l1))

  l2_roc_value = roc_auc_score(y_test, y_pred_probs_l2)
  print("l2 roc_value -------" ,l2_roc_value)
  fpr, tpr, thresholds = metrics.roc_curve(y_test, y_pred_probs_l2)
  threshold = thresholds[np.argmax(tpr-fpr)]
  print("l2 threshold --------" , threshold)

  roc_auc = metrics.auc(fpr, tpr)
  print("ROC for the test dataset-------", roc_auc)
  plt.plot(fpr,tpr,label="Test, auc="+str(roc_auc))
  plt.legend(loc=4)
  plt.show()

  Results = Results.append(pd.DataFrame({'Methodology': Methodology,'Model': 'Logistic Regression with L2 Regularisation','F1_score':f1_score_l2,'roc_value': l2_roc_value,'threshold': threshold}, index=[0]),ignore_index= True)

  l1_roc_value = roc_auc_score(y_test, y_pred_probs_l1)
  print("l1 roc_value" ,l1_roc_value)
  fpr, tpr, thresholds = metrics.roc_curve(y_test, y_pred_probs_l1)
  threshold = thresholds[np.argmax(tpr-fpr)]
  print("l1 threshold" , threshold)

  roc_auc = metrics.auc(fpr, tpr)
  print("ROC for the test dataset ----------",roc_auc)
  plt.plot(fpr,tpr,label="Test, auc="+str(roc_auc))
  plt.legend(loc=4)
  plt.show()

  Results = Results.append(pd.DataFrame({'Methodology': Methodology,'Model': 'Logistic Regression with L1 Regularisation','F1_score': f1_score_l1,'roc_value': l1_roc_value,'threshold': threshold}, index=[1]),ignore_index= True)
  return Results






















```

```python
#create a function to fit and predict SVM model
def SVMModels(Results,Methodology,X_train,y_train,X_test,y_test):
  from sklearn.svm import SVC
  from sklearn.metrics import roc_auc_score
  clf=SVC(kernel='rbf' , random_state=42)
  clf.fit(X_train,y_train)
  score=clf.score(X_test,y_test)
  print("***********Model score***********" )
  print(score)
  y_pred_SVM = clf.predict(X_test)
  SVM_f1_score=metrics.f1_score( y_true=y_test,y_pred=y_pred_SVM)
  print("f1_score " , SVM_f1_score)
  print("*******Confusion Matrix*********")
  plot_confusion_matrix(y_test,y_pred_SVM)
  print("**********Classification Report*********")
  print(classification_report(y_test,y_pred_SVM))

  #Run Classifier

  classi=SVC(kernel='rbf',probability=True)
  SVM_probs= classi.fit(X_train,y_train).predict_proba(X_test)[:,1]

  #Calcul roc auc
  roc_value =roc_auc_score(y_test,SVM_probs)
  print("SVM roc_value " , roc_value)
  fpr,tpr,thresholds = metrics.roc_curve(y_test,SVM_probs)
  threshold=thresholds[np.argmax(tpr-fpr)]
  print("Random threshold : {0}".format(threshold))
  roc_auc=metrics.auc(fpr,tpr)
  print("ROC for the test dataset" , '{:.1%}'.format(roc_auc))
  plt.plot(fpr,tpr,label="test, auc="+str(roc_auc))
  plt.legend(loc=4)
  plt.show()
  Results=Results.append(pd.DataFrame({'Methodology':Methodology , 'Model': 'SVM' , 'F1_score':SVM_f1_score , 'roc_value':roc_value , 'threshold':threshold},index=[5]),ignore_index= True)
  return Results

















```

```python
#Create a function to fit and predict KNN model
def knnmodel (Results,Methodology,X_train,y_train,X_test,y_test):
  from sklearn.metrics import roc_auc_score
  from sklearn.neighbors import KNeighborsClassifier
  from sklearn.model_selection import RandomizedSearchCV
  param_dist = {
      'n_neighbors' : list(range(1,30)) ,
      'weights': ['uniform', 'distance'] ,
      'metric' : ['euclidean' , 'manhattan' , 'chebyshev' , 'minkowski' ] ,
      'p': [1, 2] ,
      'algorithm' : ['auto', 'brute' , 'kd_tree' , 'ball_tree' ]}

  random_search = RandomizedSearchCV(
    KNeighborsClassifier() ,
    param_distributions=param_dist ,
    n_iter=50 ,
    scoring='f1' ,
    cv=5
  )
  random_search.fit(X_train, y_train)
  best_params = random_search.best_params_
  score=random_search.score(X_test,y_test)
  print("**********Model score*********" )
  print(score)

  #F1_score
  y_pred = random_search.predict(X_test)
  KNN_f1_score = metrics.f1_score(y_pred=y_pred , y_true=y_test)
  print("Confusion Matrix")
  plot_confusion_matrix(y_test,y_pred)
  print("********classification Report*******"  )
  print(classification_report(y_test,y_pred))
  knn_probs= random_search.predict_proba(X_test)[:,1]

  #Calculate rac auc
  knn_roc_value =roc_auc_score(y_test,knn_probs)
  print("KNN roc_value " , knn_roc_value)
  fpr,tpr,thresholds = metrics.roc_curve(y_test,knn_probs)
  threshold=thresholds[np.argmax(tpr-fpr)]
  print("KNN threshold : {0}".format(threshold))
  roc_auc=metrics.auc(fpr,tpr)
  print("ROC for the test dataset" , '{:.1%}'.format(roc_auc))
  plt.plot(fpr,tpr,label="test, auc="+str(roc_auc))
  plt.legend(loc=4)
  plt.show()
  Results=Results.append(pd.DataFrame({'Methodology':Methodology , 'Model': 'KNN' , 'F1_score':KNN_f1_score , 'roc_value':knn_roc_value , 'threshold':threshold},index=[3]),ignore_index= True)
  return Results

































```

```python
def votingclassmodel (Results,Methodology,X_train,y_train,X_test,y_test):
  from sklearn.ensemble import VotingClassifier
  from sklearn.tree import DecisionTreeClassifier
  from sklearn.svm import SVC
  from sklearn.metrics import roc_auc_score


  classifier1 = DecisionTreeClassifier(random_state=42)
  classifier2 = SVC(probability=True, random_state=42)
  voting_classifier = VotingClassifier(estimators=[('dt', classifier1), ('svc', classifier2)], voting='soft')
  voting_classifier.fit(X_train, y_train)
  y_pred = voting_classifier.predict(X_test)
  score=voting_classifier.score(X_test,y_test)
  print("Model score" , score)
  y_pred_SVM =voting_classifier.predict(X_test)
  VC_Accuracy=metrics.accuracy_score(y_pred=y_pred,y_true=y_test)
  VC_f1_score = metrics.f1_score(y_pred=y_pred , y_true=y_test)

  print("accuracy_score :{0}".format(VC_Accuracy))
  print("Confusion Matrix")
  plot_confusion_matrix(y_test,y_pred)
  print("Classification Report")
  print(classification_report(y_test,y_pred))
  Random_probs= voting_classifier.predict_proba(X_test)[:,1]
  Random_roc_value =roc_auc_score(y_test,Random_probs)
  print("Random roc_value : {0}".format(Random_roc_value))
  fpr,tpr,thresholds = metrics.roc_curve(y_test,Random_probs)
  threshold=thresholds[np.argmax(tpr-fpr)]
  print("Random threshold : {0}".format(threshold))
  roc_auc=metrics.auc(fpr,tpr)
  print("ROC for the test dataset" , '{:.1%}'.format(roc_auc))
  plt.plot(fpr,tpr,label="test, auc="+str(roc_auc))
  plt.legend(loc=4)
  plt.show()
  Results=Results.append(pd.DataFrame({'Methodology':Methodology , 'Model': 'VotingClassifier' , 'F1_score':VC_f1_score , 'roc_value':Random_roc_value , 'threshold':threshold},index=[4]), ignore_index= True)
  return Results

























```
